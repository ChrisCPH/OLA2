```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
```

```python
data = pd.read_csv('Housing.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Using one hot encoding to convert the data

data_fixed = pd.get_dummies(data, columns=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'])
```

```python
# Using boxplot to find outliers
sns.boxplot(data_fixed['price'])
```

```python
data_filtered = data_fixed[data_fixed['price'] <= 0.87e+07]
```

```python
sns.boxplot(data_filtered['price'])
```

```python
data_filtered.head()
```

```python
data_filtered.describe()
```

```python
data_filtered.info()
```

```python
# Splitting up the data so y is the price and X is the rest of the data

X = data_filtered.drop('price', axis=1)
y = data_filtered['price']
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting up training and testing data so that 30% of the data will be for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
modelLR = LinearRegression()
modelLR.fit(X_train, y_train)
y_predLR = modelLR.predict(X_test)
print("Linear Regression MSE: ", mean_squared_error(y_test, y_predLR))
print("Linear Regression R2 score: ", r2_score(y_test, y_predLR))
```

```python
modelRFR = RandomForestRegressor(random_state=42)
modelRFR.fit(X_train, y_train)
y_predRFR = modelRFR.predict(X_test)
print("Random Forest Regressor MSE: ", mean_squared_error(y_test, y_predRFR))
print("Random Forest Regressor R2 score: ", r2_score(y_test, y_predRFR))
```

```python
modelSVR = SVR()
modelSVR.fit(X_train, y_train)
y_predSVR = modelSVR.predict(X_test)
print("SVR MSE: ", mean_squared_error(y_test, y_predSVR))
print("SVR R2 score: ", r2_score(y_test, y_predSVR))
print("SVR Accuracy score: ", accuracy_score(y_test, y_predSVR))
```

Looking at the r2 score we can see that the linear regression model is the best on with a score of 0.62. This might not seem like a high score but when it comes to house prices theres a few important factors that are not included in my data such as location and therefore it will be harder for the model to estimate the price. This is also why we see a MSE of 1 trillion. The data also has quite a small size so it might be better to find houseprice data that includes location and has more samples.

Something else interesting is that our SVR model has a negative r2 score which means that the model is worse at predicting house score than if you used the average as the prediction.

Explain what ‘overfitting’ means and how you can spot it:

Overfitting is when the model learns the training data too well and you will get very high accurracy when using the training data but the model will not perform well on new data. This why we split data into training data and test data so that the model doesn't see all the data.
